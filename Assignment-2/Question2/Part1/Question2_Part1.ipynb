## There are three categories of json files in this folder.They are identified by the key called 'term' in each of the json file.
## Create a folder structure to read all these json files & store them into these separate folders.You are expeacted to create a hierarchy of folder structure.

```python
import json                                                  #importing json library
from glob import glob                                        #importing glob library
count=0                                                      #maintaining a counter to verify it checked all the files
list_terms = [];                                             #empty list to store 3 categories identified by key "term"
for filename in glob('data/*.json'):                         #iterrating over all the json files in data folder
    with open(filename) as r:                                #opening each files and giving alias as r
        json_data = json.load(r)                             #loading the data of each file into json_data
        count+=1                                             #increasing the counter after reading each file
        if json_data['term'] not in list_terms:              #to find unique terms,checking if the term is already in list_terms
            list_terms.append(json_data['term'])             #if the term is not in the list,it is added to the list
print("Total no. of json files for analysis "+str(count))    #printing total number of JSON files read
print(list_terms)                                            #printing the categories found
```

```python
import os                                                    #importing os

def makemydir(path):                                         #creating a function to create a folder
    if not os.path.exists(path):                             #condition to check if the folder doesn't exists already
        os.makedirs(path)                                    #folder will be created if the condition is true
        print(path+" directory created successfully")
    else:
        print(path+" already exists")                        #else folder already exists will be printed

dataProcessed_path = 'Assignment-2/DataProcessed'            #base path 
makemydir(dataProcessed_path)                                #creating the first folder
```

```python
#Creating folders for countries
countries=[]                                                 #empty list of countries
for filename in glob('data/*.json'):                         #iterrating over all the json files in data folder
     with open(filename) as r:                               #opening each files and giving alias as r
        json_data = json.load(r)                             #loading the data of each file into json_data
        location = json_data['location']                     #stroing the value of location key in a variable

        if location['country'] not in countries:             #loop for finding unique countries & creating their subfolders
            countries.append(location['country'])
            makemydir(dataProcessed_path+'/'+location['country'])
            
print(countries)                                             #printing unique countries
```

```python
from shutil import copyfile

#creating subfolders for states in countries
for i in range(0,len(countries)):                            #for every country in countries' list find states
    states=[];                                               #empty list of states for every country
    for filename in glob('data/*.json'):
        with open(filename) as r:                               
            json_data = json.load(r)
            location = json_data['location']
            if location['country']==countries[i]:
                 if location['state'] not in states:         #loop for finding unique states for a coutry & storing them
                    states.append(location['state'])
    print(countries[i])
    print(states)
    for state in states:                                     #for all the states in a country create a subfolder for each state
        makemydir(dataProcessed_path+'/'+countries[i]+'/'+state)
    
    #creating subfolders for cities in a state in a country
    for j in range(0,len(states)):
        cities=[];                                           #empty list ofcities for a state of every country
        for filename in glob('data/*.json'):
            with open(filename) as r:                               
                json_data = json.load(r)
                location = json_data['location']
                if location['country']==countries[i]:
                    if location['state']==states[j]:
                        if location['city'] not in cities:   #loop for finding unique cities for a state & storing them
                            cities.append(location['city'])
        print(countries[i])
        print(states[j])
        print(cities)
        for city in cities:                                  #for all cities,created 3 folders(attractions,restaurants,hotels)
            makemydir(dataProcessed_path+'/'+countries[i]+'/'+states[j]+'/'+city)
            makemydir(dataProcessed_path+'/'+countries[i]+'/'+states[j]+'/'+city+'/'+list_terms[0])
            makemydir(dataProcessed_path+'/'+countries[i]+'/'+states[j]+'/'+city+'/'+list_terms[1])
            makemydir(dataProcessed_path+'/'+countries[i]+'/'+states[j]+'/'+city+'/'+list_terms[2])
            
```
